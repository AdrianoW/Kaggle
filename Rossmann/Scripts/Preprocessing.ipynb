```python
%pylab inline
import mylib.utils as ut
from mylib.utils import print_time
import mylib.preprocess as pp
import pandas as pd
```

```python
# read the input files
train_raw = pd.read_csv('./Input/train.csv', parse_dates=[2])
test_raw = pd.read_csv('./Input/test.csv', parse_dates=[3])
store = pd.read_csv('./Input/store.csv')
#del test_raw['Id']
```

```python
# date processing
def date_processing(in_data):
    # create day of year, week of year, month, week of month
    data = in_data[['Date']].copy()
    data['year'] = data.Date.dt.year
    data['month'] = data.Date.dt.month
    data['day'] = data.Date.dt.day
    #data['dayofyear'] = data.Date.dt.dayofyear
    data['weekofyear'] = data.Date.dt.weekofyear
    data['quarter'] = data.Date.dt.quarter
    del data['Date']
    return data
    
# dummies of StateHoliday and their interaction
def holiday_processing(in_data):
    data = in_data[['StateHoliday']]
    #binarizer = pp.Binarizer(['StateHoliday']).fit(data)
    new_cols = pd.get_dummies(data, prefix='StateHoliday')
    return new_cols

# create dummies of day of week
def dayofweek_processing(in_data):
    data = in_data[['DayOfWeek']].copy()
    data['DayOfWeek'] = data['DayOfWeek'].astype(str)
    new_cols = pp.Binarizer(['DayOfWeek']).fit_transform(data)
    return new_cols

def fix_data_problems(in_data):
    # so far, only nan in Open, and they are all regular business dates
    #in_data.loc[data.Store ==622][data.weekofyear >=36][data.weekofyear <=39]
    in_data.loc[in_data.Open.isnull(), 'Open'] = 1
    
    # state holidays have mixed types. 0 as int or string
    in_data.loc[in_data.StateHoliday==0, 'StateHoliday'] = '0'
    
    return in_data

# create dummies of day of week
def store_types_products_processing(in_data):
    data = in_data[['StoreType', 'Assortment']].copy()
    new_cols = pp.Binarizer(['StoreType', 'Assortment']).fit_transform(data)
    return new_cols

def process_data(in_data):
    '''
    process the data, creating new columns and cleaning problems
    '''
    # clone data to not destroy old one
    data = in_data.copy()

    # fix data problems and create date fields
    print_time('Fixing data problems')
    fix_data_problems(data)
    #data = pd.concat([data, date_processing(data)], axis=1)
    print_time(data.shape)

    # create date fields
    print_time('Creating dates fields')
    tmp = date_processing(data)
    data = pd.concat([data, tmp], axis=1)
    del data['Date']
    print_time(data.shape)

    # create holiday dummies
    #print_time('Holiday dummies')
    #tmp = holiday_processing(data)
    #data = pd.concat([data, tmp], axis=1)
    #del data['StateHoliday']
    #print_time(data.shape)

    # create dayofweek dummies
    #print_time('Dummy day of week')
    #tmp = dayofweek_processing(data)
    #data = pd.concat([data, tmp], axis=1)
    #del data['DayOfWeek']
    #print_time(data.shape)
    
    # store processing
    
    # join with train and test
    print_time('Joining store')
    data = pd.merge(data, store, on='Store')
    print_time(data.shape)
    data.fillna(0, inplace=True)
    
    # create dummies for store type and product type
    print_time('Dummy store type and product assortment')
    tmp = store_types_products_processing(data)
    data = pd.concat([data, tmp], axis=1)
    del data['StoreType']
    del data['Assortment']
    print_time(data.shape)
    
    # calculate days since competition opening (maybe split into bins)
    #print_time('Calculating time since competition opening')
    #tmp = data[['Date', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear']].\
    #    apply(lambda l: make_competition_date(l), axis=1)
    #tmp_idx = tmp != -1
    #tmp[tmp_idx] = data[tmp_idx].Date-pd.to_datetime(tmp[tmp_idx])
    #tmp.name = 'competition_age_days'
    #data = pd.concat([data, tmp], axis=1)
    #print_time(data.shape)
    
    return data
```

```python
# process data for train and test datasets
train = process_data(train_raw)
test = process_data(test_raw)

# add to test the features fields that are missing 
cols_features = list(set(train.columns) - set(['Sales', 'Customers']))
cols_diff = list( set(train.columns) -set(test.columns) - set(['Sales', 'Customers']))
for cols in cols_diff:
    test[cols] = 0
```

```python
# transform column names to alphanumeric (xgboost default)
train.columns = [t.replace('_', '') for t in train.columns.tolist()]
test.columns = [t.replace('_', '') for t in test.columns.tolist()]      
```

```python
cols_features = list(set(train.columns) - set(['Sales', 'Customers', 'PromoInterval', 'StateHoliday']))
```

```python
print_time("Consider only open stores for training. Closed stores wont count into the score.")
train = train[train["Open"] != 0]
print_time("Use only Sales bigger then zero. Simplifies calculation of rmspe")
train = train[train["Sales"] > 0]
print_time(train.shape)
```

```python
train[cols_features].dtypes
```

```python
from sklearn.cross_validation import train_test_split
import xgboost as xgb

print('training data processed')

params = {"objective": "reg:linear",
          "eta": 0.01,
          "max_depth": 10,
          "subsample": 0.95,
          "colsample_bytree": 0.8,
          "seed": 3343
          }


print("Train a XGBoost model")
X_train, X_valid = train_test_split(train, test_size=0.2, random_state = 4889)
y_train = X_train.Sales
y_valid = X_valid.Sales
dtrain = xgb.DMatrix(X_train[cols_features], y_train)
dvalid = xgb.DMatrix(X_valid[cols_features], y_valid)
```

```python
num_boost_round = 3000
watchlist = [(dvalid, 'eval'), (dtrain, 'train')]
gbm = xgb.train(params, dtrain, num_boost_round, evals=watchlist, early_stopping_rounds=40, \
  feval=rmspe_xg, verbose_eval=True)
```

```python
print("Validating")
yhat = gbm.predict(xgb.DMatrix(X_valid[cols_features]))
error = rmspe(X_valid.Sales.values, yhat)
print('RMSPE: {:.6f}'.format(error))
```

```python
print("Make predictions on the test set")
dtest = xgb.DMatrix(test[cols_features])
test_probs = gbm.predict(dtest)
```

```python
# Make Submission
result = pd.DataFrame({"Id": test["Id"], 'Sales': test_probs})
result.to_csv("xgboost_nolog_submission.csv", index=False)
```
